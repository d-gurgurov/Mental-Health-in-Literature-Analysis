```python
author = 'jw_goethe'
df_ch = pd.read_csv(f'{author}.csv')
df_ch.head()
```

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download("stopwords")
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
df_ch["description_words"] = df_ch["description"].str.split()
df_ch['description_words'] = df_ch['description_words'].apply(lambda x: [lemmatizer.lemmatize(word.translate(str.maketrans('', '', string.punctuation)).lower()) for word in x if word.lower() not in stop_words])
df_ch['description_words'] = df_ch['description_words'].apply(lambda words: ' '.join(words))
```

```python
from sentence_transformers import SentenceTransformer

name = "stsb-bert-large"
model = SentenceTransformer(name)
```

```python
characters = list(df_ch.description_words.values)
characters_embeddings = model.encode(characters)
```

```python
df_ch['Embeddings'] = characters_embeddings.tolist()
```

```python
df_ch.to_csv(f'{author}_embedded.csv', index=False)
```

```python
char_names = []
for name in df_ch["name"]:
    char_names.append(name)
```

```python
import plotly.express as px
import pandas as pd
from sklearn.decomposition import PCA

# performing dimensionality reduction using PCA
pca = PCA(n_components=3)
embeddings_3d = pca.fit_transform(characters_embeddings)

# creating a DataFrame for plotly
data_df = pd.DataFrame(embeddings_3d, columns=['PC1', 'PC2', 'PC3'])
data_df['Character'] = char_names

# creating an interactive 3D scatter plot
fig = px.scatter_3d(data_df, x='PC1', y='PC2', z='PC3', text='Character')
fig.show()
```
